```python
import pandas as pd
df = pd.read_csv("D:/ML_OPS/Project-2/data/StudentsPerformance.csv")

df.head()
```

```python
# Repalcing the columns names
df.columns =df.columns.str.replace(" ","_")
df.columns =df.columns.str.replace("/","_")

```

```python
# Rename columns
df.rename(columns={
    "gender": "gender",
    "race/ethinicity": "race_ethinicity",
    "parental level of education": "parental_level_of_education",
    "lunch" : "lunch",
    "test preparation course":"test_preparation_course",
    "math score":"math_score",
    "reading score":"reading_score",
    "writing score":"writing_score"
}, inplace=True)

# Save to a new CSV
df.to_csv("D:/ML_OPS/Project-2/students_new_performance.csv", index=False)
```
